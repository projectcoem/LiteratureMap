```python
import json


with open('./src/static/authorLinksSmallerShortCleaned_new.json', 'r', encoding='utf-8') as file:
    data = json.load(file)

story_uuids=[]
for i in data["nodes"]:
    story_uuids.append(i["story_id"])
    
with open('./src/static/authorLinksSmaller_new.json', 'r', encoding='utf-8') as file:
    data2 = json.load(file)
    
dataset=set(story_uuids)
for i in data2["nodes"]:
    if i["story_id"] not in dataset:
        story_uuids.append(i["story_id"])
    
print(story_uuids)
print(len(story_uuids))
```

```python
import os

directory="./src/static/audios_es"
all_files = os.listdir(directory)
filname=all_files[0].split(".")[0]
print(filname)
for file_name in all_files:
    filname=file_name.split(".")[0]
    # Check if the file name is not in the list of story_ids
    if filname not in story_uuids:
        # Construct full file path
        file_path = os.path.join(directory, file_name)
        # Check if it is a file (not a directory)
        if os.path.isfile(file_path):
            # Remove the file
            os.remove(file_path)
            print(f"Removed: {file_path}")

print("Cleanup complete.")
directory="./src/static/audios_es"
all_files = os.listdir(directory)
print(len(all_files))
```
